```python
import pandas as pd

a = {"x": 4, "y": 2, "z": 5}
b = {"x": -2, "y": 3, "z": -1}
A = pd.Series(data=a, index=["x", "y", "z"])
B = pd.Series(data=b, index=["x", "y", "z"])

dist = ((B.x-A.x)**2 + (B.y-A.y)**2 + (B.y-A.y)**2)**0.5
dist
```
