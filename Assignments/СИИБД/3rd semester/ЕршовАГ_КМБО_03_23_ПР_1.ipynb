```python
import numpy as np
from matplotlib import pyplot as plt

arr = np.random.uniform(low=0, high=1, size=100)
print(f"Average: {np.average(arr)}, Median: {np.median(arr)}")
print("Медиана и среднее арифм. значения полученных значений очень близки => распределение симметрично.")

plt.plot(arr, "o")
plt.show()
```

```python
xx = np.arange(1, 11, 1)
yy = np.sqrt(1 + np.exp(np.sqrt(xx)) + np.cos(xx**2)) / np.abs(1-np.sin(xx)**3) + np.log(np.abs(2*xx))

plt.plot(xx, yy)
plt.show()

xx1 = xx[:int(len(xx)/2)]
yy1 = yy[:int(len(yy)/2)]

plt.scatter(xx1, yy1)
plt.show()
```

```python
xx = np.arange(0, 11, 1)
yy = np.abs(np.cos(xx*np.exp(np.cos(xx) + np.log(xx+1))))

plt.plot(xx, yy)
plt.fill_between(x=xx,y1=yy)
plt.show()

print(f"Area under plot: {np.trapz(yy)}")
```

```python
xx = np.arange(1, 13, 1)
google = np.array([87.88, 92.68, 102.83, 104.90, 120.14, 121.10,
                   124.85, 135.48, 145.65, 133.55, 148.16, 144.21])
apple = np.array([133.52, 133.75, 123.75, 123.66, 132.04, 125.08,
                  136.60, 146.36, 152.83, 141.90, 148.99, 167.48])
microsoft = np.array([335.13, 331.36, 282.12, 302.87, 286.36, 269.61,
                      251.23, 253.40, 238.47, 235.90, 235.06, 222.53])

plt.plot(xx, apple, label="Apple")
plt.plot(xx, google, label="Google")
plt.plot(xx, microsoft, label="Microsoft")
plt.legend()
plt.xlabel("Stocks")
plt.show()
print("Акции Google и Apple за 2021 год в среднем выросли, в то время как акции Microsoft упали в цене.")
print("Для улучшения визуализации результатов можно использовать другой тип графика, к примеру с отображением выбросов/медианного значения.")
```

```python
from math import e, sin, cos

print("""
Выберете операцию/функцию для вычисления (введите номер):
1. x + y
2. x - y
3. x * y
4. x / y
5. e^(x+y)
6. sin(x+y)
7. cos(x+y)
8. x^y
""")

operation = int(input())
if operation not in range(1, 9):
  print(f"Неопознанный номер операции: {operation}.")

x = float(input("Введите x:"))
y = float(input("Введите y:"))

if operation == 1:
  print(f"Результат вычисления x+y: {x+y}")
elif operation == 2:
  print(f"Результат вычисления x-y: {x-y}")
elif operation == 3:
  print(f"Результат вычисления x*y: {x*y}")
elif operation == 4:
  print(f"Результат вычисления x/y: {x/y}")
elif operation == 5:
  print(f"Результат вычисления e^(x+y): {e**(x+y)}")
elif operation == 6:
  print(f"Результат вычисления sin(x+y): {sin(x+y)}")
elif operation == 7:
  print(f"Результат вычисления cos(x+y): {cos(x+y)}")
elif operation == 8:
  print(f"Результат вычисления x^y: {x**y}")
else:
  print(f"Неопознанный номер операции: {operation}.")
```
